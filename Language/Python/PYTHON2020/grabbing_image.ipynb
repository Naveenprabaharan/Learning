```python
import requests
```

```python
res=requests.get('https://en.wikipedia.org/wiki/Mahatma_Gandhi')
```

```python
import bs4
```

```python
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
picture=soup.select('.image')[0]
```

```python
picture
```
